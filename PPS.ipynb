```python
import pandas as pd
import numpy as np
import ppscore as pps
```

```python

df = pd.DataFrame()
df["x"] = np.random.uniform(-2, 2, 1_000_000)
df["error"] = np.random.uniform(-0.5, 0.5, 1_000_000)
df["y"] = df["x"] * df["x"] + df["error"]
```

```python
df
```

```python
pps.score(df, "x", "y")
```

```python
bill = pd.read_csv('Bill_HSR.csv')
```

```python
Bill = bill.iloc[0:9999, :]
```

```python
Bill.head()
```

```python
pps.score(Bill, "TOTAL_FIXED_CHARGES", "TAX")
```

```python
pps.predictors(Bill, "TAX")
```

```python
pps.matrix(Bill)
```

```python
import seaborn as sns
predictors_df = pps.predictors(Bill, y="TAX")
sns.barplot(data=predictors_df, x="x", y="ppscore")
```

```python
matrix_df = pps.matrix(Bill)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
```

```python
sns.heatmap(matrix_df, vmin=0, vmax=1, cmap="Blues", linewidths=1, annot=True)
```
